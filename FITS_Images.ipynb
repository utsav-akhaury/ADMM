```python
import numpy as np
import numpy.fft as fft
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
data1 = fits.open('psf2asec-96.psf.fits')[0].data
data2 = fits.open('psf5asec-96.psf.fits')[0].data
print(data1.shape)
print(data2.shape)
```

```python
img1 = data1[0,0,:,:]
img2 = data2[0,0,:,:]
print(img1.shape)
print(img2.shape)
```

### Image 1  -  psf2asec

```python
plt.imshow(img1)
plt.show()
```

### Image 2  -  psf5asec

```python
plt.imshow(img2)
plt.show()
```

### Centered FFT of Image 1

```python
img1_fft = np.abs(fft.fftshift(fft.fft2(img1)))
plt.imshow(img1_fft)
plt.show()
```

### Centered FFT of Image 2

```python
img2_fft = np.abs(fft.fftshift(fft.fft2(img2)))
plt.imshow(img2_fft)
plt.show()
```

### IFFT of FFT(Image 1) compared with Image 1 (Original)

```python
img1_ifft = np.abs(fft.ifftshift(fft.ifft2(img1_fft)))
plt.subplot(1,2,1)
plt.imshow(img1_ifft)
plt.title('IFFT of FFT(Image 1)')
plt.subplot(1,2,2)
plt.imshow(img1)
plt.title('Image 1 (Original)')
plt.show()
```
